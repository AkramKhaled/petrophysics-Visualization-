# Customising the Log Plot


```python
fig,ax=plt.subplots(figsize=(10,10))

ax1=plt.subplot2grid((1,3) , (0,0) , rowspan=1 , colspan=1)
ax2=plt.subplot2grid((1,3) , (0,1) , rowspan=1 , colspan=1)
ax3=plt.subplot2grid((1,3) , (0,2) , rowspan=1 , colspan =1)
ax4=ax2.twiny() #twin the denisthy axes with neutron axes



ax1.plot("GR" , "DEPTH" , data=df2 , color="green")
ax1.set_xlabel("Gamma")
ax1.xaxis.label.set_color("green")
ax1.set_xlim(0,200)
ax1.set_ylabel("Depth")
ax1.set_ylim(4850 , 4600)
ax1.set_xticks([0 , 50 , 100 , 150 , 200 ] , color="green")
ax1.tick_params(axis="x" , color="green")
ax1.spines["top"].set_edgecolor("green")
ax1.xaxis.set_ticks_position("top")
ax1.xaxis.set_label_position("top")
ax1.spines["top"].set_position(("axes", 1.02))
ax1.grid()

ax2.plot("RHOB" , "DEPTH" , data=df2 , color="purple")
ax2.set_xlim(1.95,2.95) # set the x axis limit 
ax2.set_xlabel("Denisty") #set the title of label 
ax2.xaxis.label.set_color("purple")
ax2.xaxis.set_label_position("top") #set the position of label 
ax2.set_xticks([1.95, 2.2, 2.45, 2.7, 2.95]) #set numbers of the bar
ax2.xaxis.set_ticks_position("top") #set position of bar numbers
ax2.tick_params(axis="x" , color="purple") #set the color of bar numbers 
ax2.spines["top"].set_position(("axes" , 1.02)) #set the bar position 
ax2.spines["top"].set_edgecolor("purple") #set color of the bar
ax2.set_ylim(4850,4600) #set the Y axis limit
ax2.grid() #to make grids style on the ploting 

ax3.plot("DT" , "DEPTH" , data=df2 , color="orange")
ax3.set_xlim(140, 40)
ax3.set_xlabel("Sonic")
ax3.xaxis.label.set_color("orange")
ax3.tick_params(axis="x" , color="orange")
ax3.xaxis.set_label_position("top")
ax3.set_ylim(4850,4600)
ax3.spines["top"].set_edgecolor("orange")
ax3.spines["top"].set_position(("axes" , 1.02))
ax3.xaxis.set_ticks_position("top")
ax3.grid()

ax4.plot("NPHI" , "DEPTH" , data=df2 , color="blue")
ax4.set_xlabel("Neutron")
ax4.xaxis.set_label_position("top")
ax4.xaxis.label.set_color("blue")
ax4.set_xlim(0.45, -0.15) 
ax4.set_ylim(4850,4600)
ax4.spines["top"].set_position(("axes",1.07))
ax4.spines["top"].set_color("blue")
ax4.spines["top"].set_visible(True)
ax4.tick_params(axis="x" , color="blue")
ax4.set_xticks([0.45, 0.3, 0.15,0, -0.15])



plt.tight_layout()
```
