```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('15_9-19.csv',header=0, skiprows=[1])
```

```python
df.head()
```

```python
df.replace( -999 , np.nan , inplace=True )
```

```python
df.describe()
```

```python
fig=plt.subplots(figsize=(25,20))

ax1=plt.subplot2grid((1,5) , (0,0) , colspan=1 , rowspan=1)
ax2=plt.subplot2grid((1,5) , (0,1) , colspan=1 , rowspan=1)
ax3=plt.subplot2grid((1,5) , (0,2) , colspan=1 , rowspan=1)
ax4=plt.subplot2grid((1,5) , (0,3) , colspan=1 , rowspan=1)
ax5=ax3.twiny()

#handle X axies 
ax10=ax1.twiny()
ax10.xaxis.set_visible(False)
ax11=ax2.twiny()
ax11.xaxis.set_visible(False)
ax12=ax3.twiny()
ax12.xaxis.set_visible(False)
ax13=ax4.twiny()
ax13.xaxis.set_visible(False)
ax14=ax5.twiny()
ax14.xaxis.set_visible(False)
#plot Gamma log
ax1.plot("GR" , "DEPTH" , data=df , color="green")
ax1.set_xlabel("Gamma")
ax1.xaxis.label.set_color("green")
ax1.xaxis.set_label_position("top")
ax1.set_xlim(0,200)
ax1.set_ylim(4150,3500)

ax1.set_xticks([0, 50, 100, 150, 200])
ax1.tick_params(axis="x" , color="green")
ax1.xaxis.set_ticks_position("top")

ax1.spines["top"].set_edgecolor("green")
ax1.spines["top"].set_position(("axes" , 1.02))
ax1.grid()

#Resistivity log 
ax2.plot("RT" , "DEPTH" , data=df , color="red")
ax2.set_xlabel("Resistivity")
ax2.xaxis.label.set_color("red")
ax2.xaxis.set_label_position("top")
ax2.set_xlim(0.2, 2000)
ax2.set_ylim(4150,3500)

ax2.set_xticks([0.1, 1, 10, 100, 1000])
ax2.xaxis.set_ticks_position("top")
ax2.tick_params(axis="x" , color="red")

ax2.spines["top"].set_edgecolor("red")
ax2.spines["top"].set_position(("axes" , 1.02))
ax2.semilogx()
ax2.grid()

# Density track
ax3.plot("RHOB" , "DEPTH" , data=df , color="red")
ax3.set_xlabel("density")
ax3.xaxis.set_label_position("top")
ax3.xaxis.label.set_color("red")
ax3.set_xlim(1.95, 2.95)
ax3.set_ylim(4150,3500)

ax3.set_xticks([1.95, 2.45, 2.95])
ax3.xaxis.set_ticks_position("top")
ax3.tick_params(axis="x" , color="red")

ax3.spines["top"].set_edgecolor("red")
ax3.spines["top"].set_position(("axes" , 1.02))
ax3.grid()

#neutron track
ax5.plot("NPHI" , "DEPTH" , data=df , color="blue")
ax5.set_xlabel("Neutron")
ax5.xaxis.label.set_color("blue")
ax5.xaxis.set_label_position("top")
ax5.set_xlim(0.45, -0.15)
ax5.set_ylim(4150,3500)

ax5.set_xticks([0.45,  0.15, -0.15])
ax5.xaxis.set_ticks_position("top")
ax5.tick_params(axis="x" , color="blue")

ax5.spines["top"].set_edgecolor("blue")
ax5.spines["top"].set_position(("axes" , 1.08))
ax5.grid()

#sonic track 
ax4.plot("DT" , "DEPTH" , data=df , color="purple")
ax4.set_xlabel("Sonic")
ax4.xaxis.label.set_color("purple")
ax4.xaxis.set_label_position("top")
ax4.set_xlim(140,40)
ax4.set_ylim(4150,3500)

ax4.tick_params(axis="x" , color="purple")
ax4.xaxis.set_ticks_position("top")

ax4.spines["top"].set_edgecolor("purple")
ax4.spines["top"].set_position(("axes" , 1.02))

plt.show()
```
