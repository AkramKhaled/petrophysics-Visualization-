```python
#loading libiraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
```

```python
#loading file 
df3=pd.read_csv("L0509WellData.csv")
print(df3.head())
```

```python
df3.replace(-999 , np.nan , inplace=True)
print(df3.head())
```

```python
df3.hist( column="GR" , bins=30 , alpha=0.5 , color="grey")
```

Plotting multiple histograms on one plot


```python
col_pl = list(df3)
col_pl.remove("DEPTH")
#Setup the number of rows and columns for our plot
rows=3
cols=2

fig=plt.figure(figsize=(15,10))
for index , feature in enumerate (col_pl):
    ax=fig.add_subplot(rows,cols,index+1)
    df3[feature].hist(bins=30 , color="pink" , ax=ax)
    ax.set_title(feature+"Disteribution")
    ax.grid(color="green")
    ax.set_axisbelow(True)
    
plt.tight_layout()  
plt.show()
```

Displaying data on a crossplot (Scatterplot)


```python
df3.plot(kind="scatter" , x="NPHI" , y="RHOB" , c="GR" , colormap="YlOrRd_r" , ylim=(3,2))
plt.show()
```
