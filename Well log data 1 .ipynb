```python
#import libiraries 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
import pandas as pd 
```

```python
#Read the head of well data csv 
df=pd.read_csv('L0509WellData (1).csv')
print(df.head())
```

```python
#Read the tail of well data csv 
print(df.tail())
```

```python
#check statics before correction 
df.describe()
```

```python
#define the null values as not number 
df.replace(-999 , np.nan , inplace=True)
#And check again 
df.head()
```

```python
#check and understand of data distribution
df.hist(bins=50 , figsize=(15,10))
plt.show()
```

```python
#check skew
df.skew()
```

```python
#show data info 
df.info()
```

```python
#show all data
print(df.to_string())
```

```python
#Viewing Data on a Log Plot
df.plot(x='DEPTH' , y='GR')
plt.show()
```

```python
#scatter plot for RHOB and DRHO
df.plot(kind='scatter' , x='RHOB' , y='DRHO' )
plt.show()
```

```python
#create a colormap to plot GR , NPHI and RHOB
df.plot(kind='scatter' , x= 'NPHI' , y='RHOB' , c='GR' , colormap='jet')
plt.show()
```

```python
#create a histogrm 
df['GR'].plot(kind='hist' , bins=50 , color='r' , grid=True )
plt.show()
```

```python
#using a Kernel Density Estimate
df['GR'].plot(kind='kde' , xlim=(0,200) , color='r')
```
